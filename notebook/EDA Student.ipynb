## Adding Columns for Total Score and Average Value

```python
df['Total Score']= df['math score']+df['reading score']+df['writing score']
df['Average']=df['Total Score']/3
```

## Finding the conclusion from the table

## Finding those students who are getting the full marks in maths and different different subjects

```python
reading_full = df[df['reading score']==100]['Average'].count()
writing_full = df[df['writing score']==100]['Average'].count()
math_full = df[df['math score']==100]['Average'].count()
```

```python
reading_full,writing_full,math_full
```

```python
reading_full_20 = df[df['reading score']<=20]['Average'].count()
writing_full_20 = df[df['writing score']<=20]['Average'].count()
math_full_20 = df[df['math score']<=20]['Average'].count()
```

```python
reading_full_20,writing_full_20,math_full_20
```

## Insights from this is , that students perform well in reading and worst in maths
